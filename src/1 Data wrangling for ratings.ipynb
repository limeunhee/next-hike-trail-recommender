# Combine csv files from each trail into 1 CSV file

```python
import glob, os    
import pandas as pd

csv_files = '/Users/eunheelim/Capstone3/csv2/CA/*.csv'

df_CA = pd.concat(map(pd.read_csv, glob.glob(os.path.join('', csv_files))))
df_CA.reset_index(inplace=True)
df_CA.drop(columns=['index', 'Unnamed: 0'], inplace=True)
df_CA.to_csv('csv2/combined/CA.csv', index = False)
```
